```python
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("medicen.csv")
df.head(99)
```

```python
x = np.asarray(df[["age" , "sex" , "cp" , "trtbps" , "chol" , "fbs" , "restecg" , "thalachh" ,  "exng" , "oldpeak" , "slp" , "caa" , "thall" , "output"]])
x[0:5]
```

```python
y = np.asarray(df[["output"]])
y[0:5]
```

```python
from sklearn import preprocessing
x = preprocessing.StandardScaler().fit(x).transform(x)
x[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='newton-cg').fit(X_train,y_train)
LR
```

```python
yhat = LR.predict(X_test)
yhat
```

```python
yhat_prob = LR.predict_proba(X_test)
yhat_prob
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test, yhat,pos_label=0)
```
