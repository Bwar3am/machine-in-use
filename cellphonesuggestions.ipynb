```python
import pandas as pd
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
```

```python
cellphones_df = pd.read_csv('items.csv')
rewies_df = pd.read_csv('reviews.csv')
rewies_df.head()
```

```python
cellphones_df.head()
```

```python
nan_counts = cellphones_df.isna().sum()
print("NaN values per column:")
print(nan_counts)
```

```python
nan_counts = rewies_df.isna().sum()
print("NaN values per column:")
print(nan_counts)
```

```python
total_nan = cellphones_df.isna().sum().sum()
print(f"Total NaN values in the DataFrame: {total_nan}")
```

```python
total_nan = rewies_df.isna().sum().sum()
print(f"Total NaN values in the DataFrame: {total_nan}")
```

```python
cellphones_df = cellphones_df.dropna()

print("DataFrame after removing rows with NaN values:")
```

```python
cellphones_df = cellphones_df.drop([ 'url' , 'image' , 'reviewUrl' ], axis=1)
cellphones_df
```

```python
rewies_df = rewies_df.drop([ 'name' , 'date' , 'verified' , 'title' , 'body' ], axis=1)
rewies_df
```

```python
merged_df = pd.merge(cellphones_df , rewies_df , on='asin', how='inner')

print("Merged DataFrame:")
merged_df
```
