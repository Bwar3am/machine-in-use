```python
cellphones_df = cellphones_df.dropna()

print("DataFrame after removing rows with NaN values:")
```

```python
cellphones_df = cellphones_df.drop([ 'url' , 'image' , 'reviewUrl' , "originalPrice" ], axis=1)
cellphones_df
```

```python
rewies_df = rewies_df.drop([ 'name' , 'date' , 'verified' , 'title' , 'body'  ], axis=1)
rewies_df
```

```python
merged_df = pd.merge(cellphones_df , rewies_df , on='asin', how='inner')

print("Merged DataFrame:")
merged_df
```

```python
comparingchart = merged_df[[ "asin" ,  "rating_x", "totalReviews", "price", "rating_y" ]]
comparingchart
```

```python
user_input = userInput = [
            {'brand':'Samsung', 'rating':4.5},
            {'brand':'Apple', 'rating':5},
            {'brand':'Motorola', 'rating':4.4},
            {'brand':'Nokia', 'rating':5}
         ] 
inputcellphones = pd.DataFrame(userInput)
inputcellphones
```

```python
inputId = merged_df[merged_df['brand'].isin(inputcellphones['brand'].tolist())]
inputMovies = pd.merge(inputId, inputcellphones)
inputMovies
```
